# C001 — Gini comparado

**Afirmación a verificar (Sección 3):** *Noruega tiene un Gini ~0,27 (uno de los más bajos del mundo)*.

Este notebook explora la afirmación y compara seis países. La versión final, reproducible, está en `scripts/01_gini_comparado.py`.

```python
import sys
from pathlib import Path

# Permitir importar desde scripts/
sys.path.insert(0, str(Path.cwd().parent / 'scripts'))

import importlib
import gini_module
importlib.reload(gini_module) if 'gini_module' in dir() else None
```

```python
# Importamos directamente el módulo de análisis
import importlib.util
spec = importlib.util.spec_from_file_location('gini', '../scripts/01_gini_comparado.py')
gini = importlib.util.module_from_spec(spec)
spec.loader.exec_module(gini)

df = gini._load_data()
df
```

## Verificación rápida del claim

El texto afirma que Noruega tiene Gini ~0,27. Confirmamos:

```python
noruega = df[df['pais'] == 'Noruega'].iloc[0]
print(f"Noruega Gini = {noruega['gini']} (fuente: {noruega['fuente']}, año {noruega['anio']})")
print(f"Diferencia con afirmación '~0,27': {abs(noruega['gini'] - 0.27):.3f}")
```

```python
# Renderizamos la figura tal como aparecerá en el sitio
fig = gini.build()
fig.show()
```

## Comentarios

- El Gini de Noruega (0,276) está marginalmente por encima de "~0,27" pero la afirmación es defendible.
- Para llamarlo "uno de los más bajos del mundo" sería más riguroso compararlo con un universo de ~50+ países, no sólo seis.
- Considerar agregar Eslovenia, República Checa, Islandia que tienen Gini menor o similar para matizar la afirmación.
